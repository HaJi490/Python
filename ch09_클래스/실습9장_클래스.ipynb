```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f'Person(Name: {self.name}, Age: {self.age})'

    def __repr__(self):
        return f'Person(name={self.name!r}, age={self.age!r})'

    def __eq__(self, other):
        if isinstance(other, Person):
            return self.name == other.name and self.age == other.age
        return False

    def __len__(self):
        return len(self.name) + len(str(self.age))  # ì´ë¦„ê³¼ ë‚˜ì´ì˜ ê¸¸ì´ í•©


class Department:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f'Department(Name: {self.name})'

    def __repr__(self):
        return f'Department(name={self.name!r})'


class Employee(Person):
    def __init__(self, name, age, department):
        super().__init__(name, age)
        self.department = department

    def __str__(self):
        return f'Employee(Name: {self.name}, Age: {self.age}, Department: {self.department})'

    def __repr__(self):
        return f'Employee(name={self.name!r}, age={self.age!r}, department={self.department!r})'

    def __eq__(self, other):
        if isinstance(other, Employee):
            return super().__eq__(other) and self.department == other.department
        return False

    def __len__(self):
        return len(self.name) + len(str(self.age)) + len(self.department.name)  # ì´ë¦„, ë‚˜ì´, ë¶€ì„œì˜ ê¸¸ì´ í•©


# ë¶€ì„œ ìƒì„±
dept1 = Department("ê°œë°œ")
dept2 = Department("ì½”ë”©")

# Employee ê°ì²´ 2ê°œ ìƒì„±
emp1 = Employee("í™ê¸¸ë™", 30, dept1)
emp2 = Employee("ì´ê´‘ìˆ˜", 25, dept2)

# ê°ì²´ ì¶œë ¥
print(emp1)  # __str__ ì‚¬ìš©
print(repr(emp2))  # __repr__ ì‚¬ìš©

# ê°ì²´ ë¹„êµ
print(emp1 == emp2)  # __eq__ ì‚¬ìš©

# Person ê°ì²´ì™€ ë¹„êµ
person1 = Person("ê°•ê°ì°¬", 30)
print(emp1 == person1)  # False (ë‹¤ë¥¸ í´ë˜ìŠ¤ ë¹„êµ)
print(emp1 == emp2)  # False (ë‹¤ë¥¸ í´ë˜ìŠ¤ ë¹„êµ)
# ê¸¸ì´ ì¶œë ¥
print(len(emp1))  # Employee ê°ì²´ì˜ ê¸¸ì´
print(len(person1))  # Employee ê°ì²´ì˜ ê¸¸ì´
```

```python
# Dept í´ë˜ìŠ¤ ì •ì˜
class Dept:
    total_departments = 0  # í´ë˜ìŠ¤ ë³€ìˆ˜

    def __init__(self, name):
        self.name = name  # ë¶€ì„œ ì´ë¦„
        Dept.total_departments += 1  # ë¶€ì„œê°€ ìƒì„±ë  ë•Œë§ˆë‹¤ í´ë˜ìŠ¤ ë³€ìˆ˜ë¥¼ ì¦ê°€
    
    def show(self):
        """ë¶€ì„œ ì •ë³´ë¥¼ ì¶œë ¥"""
        print(f"ë¶€ì„œëª…: {self.name}")

    @classmethod
    def show_total_departments(cls):
        """ì „ì²´ ë¶€ì„œ ìˆ˜ë¥¼ ì¶œë ¥í•˜ëŠ” í´ë˜ìŠ¤ ë©”ì†Œë“œ"""
        print(f"ë¶€ì„œê°¯ìˆ˜: {cls.total_departments}") #clsëŠ” í´ë˜ìŠ¤ ë©”ì„œë“œê°€ í˜¸ì¶œë˜ëŠ” í´ë˜ìŠ¤ ìì²´ë¥¼ ê°€ë¦¬í‚¨ë‹¤


# Person í´ë˜ìŠ¤ ì •ì˜
class Person:
    total_persons = 0  # í´ë˜ìŠ¤ ë³€ìˆ˜

    def __init__(self, name, age):
        self.name = name  # ì´ë¦„
        self.age = age    # ë‚˜ì´
        Person.total_persons += 1  # Person ê°ì²´ê°€ ìƒì„±ë  ë•Œë§ˆë‹¤ í´ë˜ìŠ¤ ë³€ìˆ˜ë¥¼ ì¦ê°€
    
    def show(self):
        """Person í´ë˜ìŠ¤ì˜ ì •ë³´ë¥¼ ì¶œë ¥"""
        print(f"ì´ë¦„: {self.name}, ë‚˜ì´: {self.age}")

    @classmethod
    def show_total_persons(cls):
        """ì „ì²´ ì‚¬ëŒ ìˆ˜ë¥¼ ì¶œë ¥í•˜ëŠ” í´ë˜ìŠ¤ ë©”ì†Œë“œ"""
        print(f"ì‚¬ëŒìˆ˜: {cls.total_persons}")


# Employee í´ë˜ìŠ¤ëŠ” Person í´ë˜ìŠ¤ë¥¼ ìƒì†
class Employee(Person):
    total_employees = 0  # í´ë˜ìŠ¤ ë³€ìˆ˜

    def __init__(self, name, age, dept, salary):
        super().__init__(name, age)  # ë¶€ëª¨ í´ë˜ìŠ¤(Person)ì˜ ìƒì„±ì í˜¸ì¶œ
        self.dept = dept             # Dept ê°ì²´ë¥¼ ì €ì¥
        self.salary = salary         # ê¸‰ì—¬
        Employee.total_employees += 1  # Employee ê°ì²´ê°€ ìƒì„±ë  ë•Œë§ˆë‹¤ í´ë˜ìŠ¤ ë³€ìˆ˜ë¥¼ ì¦ê°€
    
    def show(self):
        """Employee í´ë˜ìŠ¤ì˜ ì •ë³´ë¥¼ ì¶œë ¥ (Person í´ë˜ìŠ¤ì˜ ì •ë³´ í¬í•¨)"""
        super().show()  # ë¶€ëª¨ í´ë˜ìŠ¤ì˜ show() í˜¸ì¶œ
        self.dept.show()  # Dept ê°ì²´ì˜ show() í˜¸ì¶œ
        print(f"ì›”ê¸‰: {self.salary}")

    @classmethod
    def show_total_employees(cls):
        """ì „ì²´ ì§ì› ìˆ˜ë¥¼ ì¶œë ¥í•˜ëŠ” í´ë˜ìŠ¤ ë©”ì†Œë“œ"""
        print(f"ì§ì›ìˆ˜: {cls.total_employees}")


# ì˜ˆì‹œ ì‚¬ìš©
dept1 = Dept("ì„¤ê³„")
dept2 = Dept("ì½”ë”©")

emp1 = Employee("í™ê¸¸ë™", 30, dept1, 50000)
emp2 = Employee("ê°•ê°ì°¬", 25, dept2, 60000)

# ê° í´ë˜ìŠ¤ì˜ ì¸ìŠ¤í„´ìŠ¤ ë©”ì†Œë“œ í˜¸ì¶œ
emp1.show()
emp2.show()

# í´ë˜ìŠ¤ ë©”ì†Œë“œ í˜¸ì¶œ
Dept.show_total_departments() #clsëŠ” Deptë¥¼ ë°›ëŠ”ë‹¤ 
Person.show_total_persons()
Employee.show_total_employees()
```

```python
# ë¦¬ìŠ¤íŠ¸ì—ì„œ ì–¸íŒ¨í‚¹
numbers = [1, 2, 3]
a, b, c = numbers  # ê° ìš”ì†Œë¥¼ a, b, cì— í• ë‹¹

print(a)  # 1
print(b)  # 2
print(c)  # 3

# íŠœí”Œì—ì„œ ì–¸íŒ¨í‚¹
coordinates = (4, 5, 6)
x, y, z = coordinates  # ê° ìš”ì†Œë¥¼ x, y, zì— í• ë‹¹

print(x)  # 4
print(y)  # 5
print(z)  # 6
```

```python
### Unzip : ê° íŠœí”Œì˜ ìš”ì†Œë“¤ì„ ë³„ë„ì˜ ë¦¬ìŠ¤íŠ¸, íŠœí”Œë¡œ ë‚˜ëˆ„ëŠ” ì‘ì—…ìœ¼ë¡œ zip()ìœ¼ë¡œ í•œë‹¤
### zip(*iterable)ìœ¼ë¡œ ì‚¬ìš©

# ì—¬ëŸ¬ íŠœí”Œì„ í¬í•¨í•œ ë¦¬ìŠ¤íŠ¸
pairs = [(1, 'a'), (2, 'b'), (3, 'c')]

# Unzip: ê°ê°ì˜ ì²« ë²ˆì§¸ ìš”ì†Œì™€ ë‘ ë²ˆì§¸ ìš”ì†Œë¥¼ ë¶„ë¦¬
numbers, letters = zip(*pairs)

print(numbers)  # (1, 2, 3)
print(letters)  # ('a', 'b', 'c')
```

```python
# ë°ì´í„° ì •ì˜
data = [(100, 'Alice'), (200, 'Bob'), (300, 'Charlie')]

# unzipìœ¼ë¡œ ë‘ ê°œì˜ ë¦¬ìŠ¤íŠ¸ë¡œ ë¶„ë¦¬
ids, names = zip(*data) ##unpackingê³¼ unzipì„ ë™ì‹œì— ì²˜ë¦¬

# ë¶„ë¦¬ëœ ë¦¬ìŠ¤íŠ¸ë¥¼ unpacking
id1, id2, id3 = ids
name1, name2, name3 = names

print(id1, name1)  # 100 Alice
print(id2, name2)  # 200 Bob
print(id3, name3)  # 300 Charlie
```

```python
### zip() ì‚¬ìš© ì˜ˆ

# ë‘ ê°œì˜ ë¦¬ìŠ¤íŠ¸
list1 = [1, 2, 3]
list2 = ['a', 'b', 'c']

# zip()ìœ¼ë¡œ ë‘ ë¦¬ìŠ¤íŠ¸ë¥¼ ë¬¶ìŒ
zipped = zip(list1, list2)

# ì¶œë ¥ (zip ê°ì²´ë¥¼ ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í™˜)
print(list(zipped))  # [(1, 'a'), (2, 'b'), (3, 'c')]
```

```python
# ì´ë¯¸ zipëœ ë¦¬ìŠ¤íŠ¸
zipped_list = [(1, 'a'), (2, 'b'), (3, 'c')]

# zip(*zipped_list)ë¡œ unzip
numbers, letters = zip(*zipped_list)

print(numbers)  # (1, 2, 3)
print(letters)  # ('a', 'b', 'c')
```

zip()ì€ ë¬¶ëŠ” ì‘ì—…ì´ê³ , unzipì€ zip(*...)ì„ í†µí•´ ë¬¶ì¸ ë°ì´í„°ë¥¼ ë‹¤ì‹œ í‘¸ëŠ” ì‘ì—…

```python
# ë‘ ë¦¬ìŠ¤íŠ¸ ì •ì˜
numbers = [1, 2, 3]
letters = ['a', 'b', 'c']

# zip() ì‚¬ìš©: ë¦¬ìŠ¤íŠ¸ë¥¼ ë¬¶ìŒ
zipped = list(zip(numbers, letters))
print(f"Zipped: {zipped}")  # [(1, 'a'), (2, 'b'), (3, 'c')]

# unzip: zipì„ ë‹¤ì‹œ í’€ì–´ì„œ ë¦¬ìŠ¤íŠ¸ë¡œ ë¶„ë¦¬
unzipped_numbers, unzipped_letters = zip(*zipped)
print(f"Unzipped numbers: {unzipped_numbers}")  # (1, 2, 3)
print(f"Unzipped letters: {unzipped_letters}")  # ('a', 'b', 'c')
```

í´ë˜ìŠ¤ ì¸ìŠ¤í„´ìŠ¤ë¥¼ ë°°ì—´ì²˜ëŸ¬ ì‚¬ìš©í•˜ê¸° 

```python
class MyList:
    def __init__(self, initial_data):
        # ì´ˆê¸° ë°ì´í„°ë¥¼ ë¦¬ìŠ¤íŠ¸ë¡œ ì €ì¥
        self.data = list(initial_data)
    
    # ì¸ë±ìŠ¤ ì ‘ê·¼ì„ ìœ„í•œ __getitem__ ë©”ì„œë“œ êµ¬í˜„
    def __getitem__(self, index): ### ë¨¸ì‹ ëŸ¬ë‹ì—ì„œ ì‚¬ìš©ë¨ 
        print("__getitem__() í˜¸ì¶œë¨")
        return self.data[index]
    
    # ì¸ë±ìŠ¤ë¥¼ í†µí•´ ê°’ì„ ì„¤ì •í•˜ëŠ” __setitem__ ë©”ì„œë“œ êµ¬í˜„
    def __setitem__(self, index, value):
        print("__setitem__() í˜¸ì¶œë¨")
        self.data[index] = value
    
    # ì¸ë±ìŠ¤ë¥¼ í†µí•´ ê°’ì„ ì‚­ì œí•˜ëŠ” __delitem__ ë©”ì„œë“œ êµ¬í˜„
    def __delitem__(self, index):
        del self.data[index]
    
    # len() í•¨ìˆ˜ê°€ ì‘ë™í•˜ë„ë¡ __len__ ë©”ì„œë“œ êµ¬í˜„
    def __len__(self):
        return len(self.data)
    
    # ì¶œë ¥í•  ë•Œ ì¸ìŠ¤í„´ìŠ¤ì˜ ë¦¬ìŠ¤íŠ¸ ë°ì´í„°ë¥¼ ë³´ì—¬ì£¼ê¸° ìœ„í•œ __repr__ ë©”ì„œë“œ êµ¬í˜„
    def __repr__(self):
        print("__repr__() í˜¸ì¶œë¨")
        return repr(self.data)

# ì‚¬ìš© ì˜ˆì‹œ
my_list = MyList([1, 2, 3, 4])
print('---ì¸ë±ìŠ¤ ì‚¬ìš©---')
# ë°°ì—´ì²˜ëŸ¼ ì ‘ê·¼
print(my_list[0])  # 1

print('---ë³€ê²½---')
# ë°°ì—´ì²˜ëŸ¼ ê°’ ë³€ê²½
my_list[1] = 10
print('---ì¶œë ¥---')
print(my_list)  # [1, 10, 3, 4]

print('---ì‚­ì œ---')
# ë°°ì—´ì²˜ëŸ¼ ê°’ ì‚­ì œ
del my_list[2]
print(my_list)  # [1, 10, 4]

# ë°°ì—´ì˜ ê¸¸ì´ í™•ì¸
print(len(my_list))  # 3
```
